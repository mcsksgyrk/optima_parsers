```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
dir_path='./1dataMin'
data_file_paths=[]
for root, dirs, files in os.walk(dir_path):
    for file in files:
        file_path = os.path.join(root, file)
        if 'csv' in file_path:
            data_file_paths.append(file_path)
        print(f'File: {file_path}')
```

```python
import re
ICs = dict()
with open('ICk_toParse', 'r') as file:
    for line in file:
        key = re.findall(r'preferredKey="(.*?)"',line)[0]
        value = re.findall(r'units="mole fraction" >(.*?)</amount>',line)[0]
        ICs[key] = float(value)
ICs['eS'] = 0
```

```python
def defineInput(fileName):
    parts = fileName.split('_')
    for p in parts:
        for i in inps:
            if i in p:
                return  
```

```python
ic_list=[]
for f in data_file_paths:
    ic_data = ICs.copy()
    df = pd.read_csv(f)
    columns = df.columns
    for c in columns:
        if c in ICs.keys():
            df[c] = df[c]*ic_data[c]
            ic_data[c] = df[c][0]
    ic_list.append(ic_data)
ic_list
```
